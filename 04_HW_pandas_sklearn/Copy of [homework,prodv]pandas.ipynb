**4).** Какую долю (вещественное число от 0 до 1, округлить до 4-го знака) от всех покупателей составляют ВМЕСТЕ мужчины от 26 до 35 лет и женщины старше 36 лет (то есть нужно учесть несколько возрастных категорий)? (речь не об уникальных ID, а о количестве таких строк)  

```python
sum(data.Gender == 'M')
```

```python
sum(data.Age == '26-35')
```

```python
male_cond = ((data.Gender == 'M') & (data.Age == '26-35'))
sum(male_cond)
```

```python
female_cond = (data['Gender'] == 'F') & data['Age'].isin(['36-45', '46-50', '51-55', '55+'])
sum(female_cond)
```

```python
num = sum(male_cond | female_cond)
```

```python
# answer to 4
round(num / data.shape[0], 4)
```
